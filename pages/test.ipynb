```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.tools import FunctionTool

def get_weather(city):
    return f"The weather in {city} is sunny with a high of 25Â°C."

weather_tool = FunctionTool.from_function(get_weather)

llm = ChatOpenAI(
    temperature=0.1,
).bind(
    function_call={"auto": True},
    functions=[weather_tool],  # Pass as a list!
)

prompt = PromptTemplate.from_template("What's the weather like in {city}")

chain = prompt | llm

response = chain.invoke({"city": "rome"})

print(response.content)
```
